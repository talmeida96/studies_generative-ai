# Visualizaes com *Matplotlib*

Nesta seo,  explorado o uso do *Matplotlib* para criar uma ampla variedade de visualizaes de dados, desde grficos bsicos at representaes mais complexas. *Matplotlib* simplifica a criao de grficos de linhas, barras, histogramas, e grficos de rea, como  visto em exemplos prticos que ilustram essas tcnicas com o *dataset* Iris. Por exemplo,  demonstrado como um grfico de linhas pode ser utilizado para visualizar o comprimento e a largura das spalas, enquanto os histogramas so usados para observar a distribuio de caractersticas como o comprimento das ptalas. Ressalta-se a flexibilidade do *Matplotlib* em ajustar detalhes visuais como legendas, ttulos e estilos de linha, permitindo que os/as usurios/as personalizem os grficos para atender s necessidades especficas de suas anlises e apresentaes de dados.

```python
# Define o tamanho da figura (largura, altura) em polegadas
plt.figure(figsize=(10, 6))


# Adiciona o comprimento das spalas ao grfico (por padro ser adicionada uma nova linha para cada plot)
plt.plot(iris['sepal_length'], label='Comprimento da Spala')

# Adiciona a largura das spalas ao grfico
plt.plot(iris['sepal_width'], label='Largura da Spala')

# Adiciona um ttulo ao grfico
plt.title('Medidas das Spalas das ris')

# Adiciona rtulos aos eixos x e y
plt.xlabel('ndice')
plt.ylabel('Comprimento em cm')

# Adiciona uma legenda ao grfico
plt.legend()

# Exibe o grfico
plt.show()
```

```python
# Agrupando os dados por espcie (species) e calculando a mdia do comprimento das spalas (sepal_length)
sepal_length_means = iris.groupby('species')['sepal_length'].mean()
sepal_length_means
```

```python
# Criando um grfico de barras, utilizando a cor ciano (cyan) e transparncia de 70% (alpha=0.7)
sepal_length_means.plot(kind='bar', color='cyan', alpha=0.7)

# Definindo os rtulos e ttulo do grfico
plt.title('Mdia do Comprimento das Spalas por Espcie')
plt.xlabel('Espcie')
plt.ylabel('Mdia do Comprimento das Spalas (cm)')
plt.show()
```

```python
# Define a cor, a transparncia (alpha) e o nmero de barras (bins)
plt.hist(iris['petal_length'], bins=20, color='green', alpha=0.7)


plt.title('Distribuio do Comprimento das Ptalas')
plt.xlabel('Comprimento das Ptalas em cm')
plt.ylabel('Frequncia')
plt.show()
```

```python
# Preparando os dados para o grfico de rea
x = range(len(iris))  # ndice para cada amostra
x
```

Alguns tipos de grficos, como os grficos de rea, so melhor representados quando no h dados nulos ou faltantes na srie. Para isso, pode-se substituir por um valor padro, por meio do mtodo `fillna`. O cdigo a seguir substitui os dados faltantes nas colunas `sepal_length` e `sepal_width`.

```python
# Comprimento e largura da spala, tratando possveis dados faltantes
y_sepal_length = iris['sepal_length'].fillna(0)
y_sepal_width = iris['sepal_width'].fillna(0)
print(y_sepal_length)
print(y_sepal_width)
```

O cdigo a seguir gera um grfico de rea (`stackplot`) com os dados j tratados em `y_sepal_length` e `y_sepal_width`. Os rtulos so definidos utilizando o parmetro `labels`, bem como as cores por meio do parmetro `colors`. Por fim,  definido um grau de transparncia (`alpha=0.5`).

```python
# Criando um grfico de rea
plt.figure(figsize=(10, 6))
plt.stackplot(x, y_sepal_length, y_sepal_width, labels=['Comprimento da Spala', 'Largura da Spala'], colors=['lightblue', 'lightgreen'], alpha=0.5)
plt.title('Distribuio Acumulada das Medidas das Spalas')
plt.xlabel('ndice da Amostra')
plt.ylabel('Medidas das Spalas (cm)')
plt.legend(loc='upper left')
plt.show()
```

```python
# Criando um grfico de disperso do Comprimento vs Largura das Ptalas
plt.scatter(iris['petal_length'], iris['petal_width'], c='red', alpha=0.5)
plt.title('Grfico de Disperso do Comprimento vs Largura das Ptalas')
plt.xlabel('Comprimento das Ptalas em cm')
plt.ylabel('Largura das Ptalas em cm')
plt.show()
```

```python
# Contando quantas amostras existem de cada espcie
species_counts = iris['species'].value_counts()
species_counts
```

No exemplo a seguir, `species_counts` representa os valores, `species_counts.index` os rtulos, e `autopct='%1.1f%%'` as porcentagens em cada segmento. A opo `startangle=90` ajusta o ngulo inicial do grfico para facilitar a visualizao, e as cores so escolhidas manualmente para cada segmento, melhorando a clareza e distino entre as categorias.


```python
# Criando um grfico de pizza com a Distribuio das Espcies de ris
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lavender'])
plt.title('Distribuio das Espcies de ris')
plt.show()
```

```python
# Exemplo de grficos mltiplos usando subplots

# Cria uma figura e dois subplots lado a lado
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

# Grfico de Disperso com o comprimento das spalas no eixo X e a largura no eixo Y
axs[0].plot(iris['sepal_length'], iris['sepal_width'], 'b^')
axs[0].set_xlabel('Comprimento da Spala (cm)')
axs[0].set_ylabel('Largura da Spala (cm)')
axs[0].set_title('Grfico de Disperso')

# Histograma do comprimento das spalas
axs[1].hist(iris['sepal_length'], bins=15, color='g', alpha=0.7)
axs[1].set_xlabel('Comprimento da Spala (cm)')
axs[1].set_ylabel('Frequncia')
axs[1].set_title('Histograma')

# Ajusta o espaamento entre os subplots
plt.tight_layout()
plt.show()
```

# Visualizaes com *Seaborn*

Esta seo aborda como *Seaborn*, uma biblioteca de visualizao de dados construda sobre o *Matplotlib*, facilita a criao de grficos estatsticos avanados e esteticamente agradveis. *Seaborn*  destacado por sua interface de alto nvel que permite aos/s usurios/as gerar visualizaes complexas, como *pair* *plots*, *joint plots* e *KDE plots*, com menos cdigo e configuraes padronizadas visualmente mais atraentes. A seo explora ferramentas especficas do *Seaborn* que so eficazes para analisar relaes entre variveis numricas e categorias, como grficos de disperso que detalham as interaes entre variveis e grficos de caixa e *violin* *plots* para comparar distribuies e identificar *outliers* entre categorias. Essas funcionalidades tornam *Seaborn* uma escolha poderosa para anlises exploratrias de dados, proporcionando *insights* profundos por meio de visualizaes que destacam padres e tendncias nos dados de maneira intuitiva e acessvel.

```python
# Criar um pair plot das variveis do dataset Iris
# - 'iris'  o DataFrame que contm os dados do conjunto Iris
# - 'hue' define a varivel para colorir os pontos (neste caso, a espcie das flores)
# - 'markers' define os smbolos usados para cada espcie: "o" para setosa, "s" para versicolor, "D" para virginica
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])

# Adicionar um ttulo ao grfico com ajuste na posio vertical (y) para evitar sobreposio
plt.suptitle('Pair Plot das Variveis do Dataset Iris', y=1.02)

# Mostrar o grfico gerado
plt.show()
```

```python
# Cria um grfico conjunto (joint plot) que mostra a relao entre o comprimento das ptalas e a largura das ptalas
# Define o eixo x como a coluna 'petal_length' do DataFrame
# Define o eixo y como a coluna 'petal_width' do DataFrame
# Especifica o DataFrame a ser usado (iris)
# Define o tipo de grfico como 'scatter' (disperso)
# Define a cor dos pontos como magenta ('m')
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='scatter', color='m')
plt.suptitle('Joint Plot de Comprimento e Largura das Ptalas', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Cria um grfico de densidade (KDE plot) para a distribuio do comprimento das spalas
# Especifica a coluna 'sepal_length' do DataFrame iris como os dados a serem plotados
# Define a cor do grfico como vermelho (r)
# Preenche a rea sob a curva
# Adiciona um rtulo ao grfico para a legenda
sns.kdeplot(iris['sepal_length'], fill=True, color="r", label="Comprimento da Spala")

# Cria um grfico de densidade (KDE plot) para a distribuio da largura das spalas
# Especifica a coluna 'sepal_width' do DataFrame iris como os dados a serem plotados
# Define a cor do grfico como azul (b)
# Preenche a rea sob a curva
# Adiciona um rtulo ao grfico para a legenda
sns.kdeplot(iris['sepal_width'], fill=True, color="b", label="Largura da Spala")

plt.title('KDE Plots de Comprimento e Largura das Spalas')
plt.xlabel('Centmetros')
plt.ylabel('Densidade')
# Adiciona uma legenda ao grfico
plt.legend()
plt.show()
```

```python
# Boxplot com a Distribuio do Comprimento das Ptalas por Espcie
plt.figure(figsize=(10, 6))

# Define o eixo x como a coluna 'species' do DataFrame
# Define o eixo y como a coluna 'petal_length' do DataFrame
# Especifica o DataFrame a ser usado (iris)
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title('Distribuio do Comprimento das Ptalas por Espcie')
plt.xlabel('Espcie')
plt.ylabel('Comprimento das Ptalas (cm)')
plt.show()
```

```python
# Violin plot com a Distribuio da Largura das Spalas por Espcie
plt.figure(figsize=(10, 6))

# Define o eixo x como a coluna 'species' do DataFrame
# Define o eixo y como a coluna 'sepal_width' do DataFrame
# Especifica o DataFrame a ser usado (iris)
# Define a paleta de cores a ser usada (Pastel1)
# Adiciona cor s violas com base na espcie (para melhor visualizao)
# Remove a legenda
sns.violinplot(x='species', y='sepal_width', data=iris, palette='Pastel1', hue='species', legend=False)

# Adiciona um ttulo e rtulos ao grfico
plt.title('Distribuio da Largura das Spalas por Espcie')
plt.xlabel('Espcie')
plt.ylabel('Largura das Spalas (cm)')

# Mostra o grfico gerado
plt.show()
```

```python
# Criando um bar plot para comparar a mdia do comprimento das spalas entre as espcies
plt.figure(figsize=(10, 6))

# Define o eixo x como a coluna 'species' do DataFrame
# Define o eixo y como a coluna 'sepal_length' do DataFrame
# Especifica o DataFrame a ser usado (iris)
# Define a paleta de cores a ser usada (Accent)
# Adiciona cor s barras com base na espcie (para melhor visualizao)
# Remove a legenda do grfico
sns.barplot(x='species', y='sepal_length', data=iris, palette='Accent', hue='species', legend=False)
plt.title('Mdia do Comprimento das Spalas por Espcie')
plt.xlabel('Espcie')
plt.ylabel('Comprimento Mdio das Spalas (cm)')
plt.show()
```
